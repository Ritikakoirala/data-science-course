Write a Python program using Pandas to create a table that contains the following columns and data:

Employee Name: John, Kate, Leo

Department: IT, HR, Sales

Salary: 50000, 55000, 60000

```python
import pandas as pd   
df=pd.DataFrame({

"Employee Name": ["John", "Kate", "Leo"],

'Department': ['IT', 'HR', 'Sales'],

'Salary': [50000, 55000, 60000]})
print(df)
```

Using a list of lists, create a Pandas DataFrame for the following data and set the column name as "Employee", "Department", "Salar

```python

columns = ["Employee Name", "Department", "Salary"]
data = [
    ["John", "IT", 50000],
    ["Kate", "HR", 55000],
    ["Leo", "Sales", 60000]
]
df = pd.DataFrame(data, columns=columns)
print(df)
```
