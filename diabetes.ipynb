```python
#Diabetes probability increases sharply with higher risk score.
df['risk_score'] = (
    (df['age'] > 45) +
    (df['bmi'] > 30) +
    (df['HbA1c_level'] > 6.5) +
    (df['blood_glucose_level'] > 140) +
    (df['heart_disease'] == 1)
).astype(int)

```

```python
df.groupby('risk_score')['diabetes'].mean() * 100
df.plot(marker='o')
plt.xlabel('Risk Score')
plt.ylabel('Diabetes Percentage')
plt.title('Diabetes Probability by Risk Score')
plt.show()

```

A composite risk score was created using clinically significant thresholds related to age, BMI, HbA1c level, blood glucose level, and heart disease status. Each factor contributed equally to the risk score. The group-wise mean of the binary diabetes variable was used to estimate diabetes probability across risk levels. The resulting plot shows a strong positive relationship between risk score and diabetes prevalence, demonstrating the cumulative effect of multiple risk factors.
