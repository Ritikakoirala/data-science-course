How do you plot a simple line graph using Matplotlib?

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

What is the difference between plt.plot() and plt.scatter()?

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x,y)
plt.show()
plt.scatter(x,y)
plt.show()
```

How do you change the line color and style in a line plot?



```python
plt.plot(x, y, color='red', linestyle='--')
```

How can you set the x-axis and y-axis labels?


```python
plt.xlabel("this is the x ")
plt.ylabel("this is the y")
```

Which function is used to create histograms?


```python
plt.hist(x,y)
plt.show()
```

How do you add a legend to a plot?


```python
plt.plot(x, y, label="Sales")
plt.legend()
```

What does the alpha parameter control?


```python
plt.scatter(x, y, alpha=0.5)
```

How do you plot multiple subplots in the same figure?



```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.scatter(x, y)
plt.show()
```

What is the difference between plt.bar() and plt.barh()?


```python
plt.subplot(1,2,1)
plt.bar(x,y)
plt.subplot(1,2,2)
plt.barh(x,y)
```

How do you save a Matplotlib figure as an image?

```python
plt.savefig("plots.png")
```
