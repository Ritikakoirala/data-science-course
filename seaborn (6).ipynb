---Different categories of plot in Seaborn 
Plots are basically used for visualizing the relationship between variables. Those variables can be either completely numerical or a category like a group, class, or division. Seaborn divides the plot into the below categories - 
 

1.Relational plots: This plot is used to understand the relation between two variables.
2.Categorical plots: This plot deals with categorical variables and how they can be visualized.
3.Distribution plots: This plot is used for examining univariate and bivariate distributions
4.Regression plots: The regression plots in Seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.
5.Matrix plots: A matrix plot is an array of scatterplots.
6.Multi-plot grids: It is a useful approach to draw multiple instances of the same plot on different subsets of the dataset.

| Feature                  | **Matplotlib**                       | **Seaborn**                                          |
| ------------------------ | ------------------------------------ | ---------------------------------------------------- |
| **Library Type**         | Base plotting library                | High-level wrapper built on Matplotlib               |
| **Purpose**              | General plotting                     | Statistical data visualization                       |
| **Ease of Use**          | Requires more code for complex plots | Shorter, simpler syntax for complex visualizations   |
| **Plot Aesthetics**      | Basic (default)                      | Beautiful by default (color themes, style)           |
| **Integration**          | Raw control over every plot element  | Integrates easily with pandas DataFrames             |
| **Specialty**            | Line, bar, scatter, etc.             | Distribution plots, categorical plots, heatmaps      |
| **Customization**        | Very flexible but verbose            | Less flexible, but integrates well with Matplotlib   |
| **Statistical Features** | Needs manual coding for stats        | Built-in statistical functions like regression lines |

The function displot() stands for "distribution plot", and it creates:

Histogram (default)


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [10, 20, 20, 30, 40, 40, 40, 50, 60]

# Basic distribution plot (default: histogram)
sns.displot(data, kde = True)  
plt.title("Distribution of Data")  # Title for the first plot

# Distribution plot with additional features
sns.displot(
    data,        # Data to be plotted
    kde=True,    # Adds a Kernel Density Estimate (KDE) curve
    bins=10,      # Number of histogram bins
    rug=True,    # Adds small marks at each observation if True
)

plt.title("Histogram with KDE and Rug")  # Title for the second plot

plt.show()  # Show plots
```

| Parameter       | What It Does                                                                                                      |
| --------------- | ----------------------------------------------------------------------------------------------------------------- |
| `kde=True`      | Adds a **smooth curved line** (called KDE: Kernel Density Estimate) that shows the **probability distribution**.  |
| `bins=5`        | Divides the data into **5 bins** (groups). Histogram bars show how many values fall in each bin.                  |
| `rug=True`      | Adds **tick marks** at the bottom of the plot  each tick represents one data point. Helps see individual values. |
| `color='green'` | Makes the bars and KDE curve **green** in color.                                                                  |

The fmri dataset contains simulated functional MRI measurements of brain responses over time. It is commonly used for plotting multiple time series grouped by conditions.

Draws a pie chart based on the counts of each brain region.

labels=region_counts.index sets the pie slice labels to brain region names.

autopct='%1.1f%%' shows the percentage of each slice on the chart (e.g., 25.4%).

startangle=140 rotates the pie chart to start from 140 degrees for better orientation.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load fMRI dataset
fmri = sns.load_dataset("fmri")

# Count number of samples per brain region
region_counts = fmri['region'].value_counts()

# Plot pie chart
# plt.figure(figsize=(6,6))  #Initializes a new figure for the pie chart.

# figsize=(6,6) makes the figure square, ensuring the pie chart is circular.
plt.pie(region_counts, labels=region_counts.index, autopct='%1.5f%%', startangle=0)
plt.title("Distribution of fMRI Samples by Brain Region")
plt.show()

# Seaborn does not provide a built-in function for pie charts. Pie charts are generally created using Matplotlib instead.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fMRI dataset
fmri = sns.load_dataset("fmri")

# Count samples per brain region
region_counts = fmri['region'].value_counts().reset_index()
region_counts.columns = ['region', 'count']
print(region_counts)

# Plot bar graph
plt.figure(figsize=(8,5))  
# figsize=(width, height) in inches; e.g., (6,4), (10,6), etc.

sns.barplot(
    data=region_counts,     # DataFrame to use for plotting
    x='region',             # Categorical variable for x-axis
    y=[50,100],    # Numeric variable for y-axis
    hue = 'region', 
    palette='deep'  , 
    orient = 'v',
    # Color palette; possible values: 'deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind'
                            # Or a list of custom colors, e.g., ['red','blue']
)

plt.title("Number of fMRI Samples by Brain Region")
# Possible additional arguments: fontsize=14, color='blue', loc='left'/'center'/'right'

plt.xlabel("Brain Region")
# Possible arguments: fontsize=12, labelpad=10, color='green'

plt.ylabel("Sample Count")
# Same arguments as xlabel

plt.show()

print(fmri["region"].value_counts())
sns.countplot(data=fmri, x='region')
```

What is sns.relplot()?
relplot stands for relational plot.

Its a figure-level function designed to visualize relationships between two variables.

Internally, it combines scatter plots and line plots, depending on parameters.

Its very flexible  great for creating scatter plots and line plots with easy faceting.

==============================================================================================

The seaborn.relplot() method of the Seaborn library is used to plot the relation between two variables of a dataset on a graph(x-y axis) with semantic mappings of subset.

sns.relplot(data=df, x='x_column', y='y_column', kind='scatter' or 'line', hue='category', style='category', col='facet_col')

----------------------------------------------------------------
data: your DataFrame

x: variable for x-axis

y: variable for y-axis

kind: "scatter" (default) or "line"

hue: groups data by color

style: groups data by marker style (only in scatter plots)

col / row: create subplots (facets) by different categories

| Parameter       | Description                                                                | Notes / Example                             |
| --------------- | -------------------------------------------------------------------------- | ------------------------------------------- |
| **data**        | The DataFrame containing your data                                         | `data=tips`                                 |
| **x**           | Column name or vector for x-axis values                                    | `x='total_bill'`                            |
| **y**           | Column name or vector for y-axis values                                    | `y='tip'`                                   |
| **kind**        | Type of plot: `"scatter"` (default) or `"line"`                            | `kind='line'`                               |
| **hue**         | Grouping variable that defines the color of plot elements                  | `hue='smoker'`                              |
| **style**       | Grouping variable that defines marker style (only for scatter plots)       | `style='time'`                              |
| **size**        | Grouping variable that defines marker size (only for scatter plots)        | `size='size'`                               |
| **palette**     | Color palette for hue groups                                               | `palette='pastel'`                          |
| **sizes**       | Tuple specifying min and max size for markers (when using size parameter)  | `sizes=(20, 200)`                           |
| **markers**     | Boolean or list of marker styles to use (when using style parameter)       | `markers=True` or `markers=['o', 's', 'D']` |
| **alpha**       | Opacity level for the plot elements (0 = transparent, 1 = opaque)          | `alpha=0.7`                                 |
| **col**         | Column name to facet the plot into multiple subplots arranged horizontally | `col='day'`                                 |
| **row**         | Column name to facet the plot into multiple subplots arranged vertically   | `row='time'`                                |
| **col\_wrap**   | Integer number to wrap facet columns into multiple rows                    | `col_wrap=2`                                |
| **height**      | Height (in inches) of each facet subplot                                   | `height=4`                                  |
| **aspect**      | Aspect ratio (width/height) of each facet subplot                          | `aspect=1.5`                                |
| **legend**      | Whether to display legend (`'auto'`, `'brief'`, `'full'`, `False`)         | `legend='full'`                             |
| **facet\_kws**  | Dictionary of additional keyword arguments passed to `FacetGrid`           | `facet_kws={"margin_titles": True}`         |
| **ci**          | Confidence interval size for line plots (if `kind="line"`)                 | `ci=95` or `ci=None`                        |
| **estimator**   | Statistical estimator for line plots (`mean` by default)                   | `estimator=np.median`                       |
| **err\_style**  | Error bar style for line plots: `"band"` or `"bars"`                       | `err_style="bars"`                          |
| **legend\_out** | Whether to draw the legend outside the plot area (boolean)                 | `legend_out=True`                           |


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Single scatter plot: Total bill vs Tip
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.title("Scatterplot: Total Bill vs Tip (Colored by Smoker)")
plt.show()
```

This creates one scatter plot.

Points are colored by smoker status (hue='smoker').

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset (contains restaurant bills, tips, and customer info)
tips = sns.load_dataset('tips')

# Create faceted scatter plots using relplot
# data=tips          -> Dataset to use
# x='total_bill'     -> Column for x-axis
# y='tip'            -> Column for y-axis
# hue='smoker'       -> Color points by smoker status (Yes/No)
# col='time'         -> Create separate plots (facets) for each meal time (Lunch/Dinner)
# style='size'       -> Different markers for different party sizes
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker', col='time', style='size')

# Add a super title above all facets
# y=1.05 moves the title slightly above the plots
plt.suptitle("Relplot: Total Bill vs Tip Faceted by Time (Lunch, Dinner)", y=1.05)

# Display the plots
plt.show()
```

Step-by-step explanation:
sns.relplot(...):
This creates a relational plot (scatter plot by default) from the tips dataset.

data=tips:
We use the tips dataset, which has restaurant bill and tip data.

x='total_bill' and y='tip':
The x-axis shows the total bill amount, and the y-axis shows the tip given.

hue='smoker':
Points are colored differently based on whether the person is a smoker or not ("Yes" or "No").

col='time':
The plot is split into multiple subplots (faceted) by the time column, which has values "Lunch" and "Dinner".
This means:

One scatter plot for lunch data

One scatter plot for dinner data

plt.suptitle(...):
Adds a main title above all subplots for the whole figure.

plt.show():
Displays the plot.

What you see as output:
Two scatter plots side-by-side:

Left plot: Tips vs total bill during Lunch

Right plot: Tips vs total bill during Dinner

Points in each plot are colored by smoker status.

Why use relplot with col?
To compare patterns between lunch and dinner easily.

Faceting helps to break data into smaller groups for better visualization.

--------------------------------------------------------
plt.suptitle() adds a main title for the entire figure, not just one subplot.

The string inside quotes is the text of the title:

"Relplot: Total Bill vs Tip Faceted by Time (Lunch, Dinner)"

The parameter y=1.05 adjusts the vertical position of the title:

y=1 means the title is right above the top edge of the figure.

y=1.05 moves the title a little higher, so it doesnt overlap with the subplots.

What you get from JointPlot:

Scatter plot (or other plot) of total_bill vs tip in the main central area.

Shows how tips relate to total bills for each record.

Histograms (by default) on the top and right margins:

The top shows distribution of total_bill values.

The right shows distribution of tip values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset (restaurant bills, tips, and customer info)
tips = sns.load_dataset('tips')

# Create a joint plot to visualize the relationship between two variables
# x = tips.total_bill   -> Column for x-axis (total bill amount)
# y = tips.tip          -> Column for y-axis (tip amount)
# By default, it creates a scatter plot with marginal histograms
sns.jointplot(x=tips.total_bill, y=tips.tip, kind="hex")

# Display the plot
plt.show()
```

kind='scatter'  scatter plot (default)

kind='reg'  scatter plot with regression line

kind='kde'  2D density plot

kind='hex'  hexbin plot for density
